Este trabajo de ciencia de datos busca analizar y comprender patrones temporales, geográficos y de forma de los avistamientos, así como explorar las posibles correlaciones entre los datos y eventos sociales o culturales.

Este análisis no solo busca proporcionar una comprensión más profunda de los avistamientos de OVNIs, sino también contribuir al debate sobre cómo estos fenómenos son percibidos, documentados y estudiados a lo largo del tiempo.

```python
# Paso 1: Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Paso 2: Cargar el dataset
df_ufo = pd.read_csv('/content/ufo-sightings.csv')

# Paso 3: Eliminar valores nulos en la columna 'Season'
df_ufo = df_ufo[df_ufo['Season'].notna()]

# Paso 4: Contar los encuentros por cada estación
encounters_per_season = df_ufo['Season'].value_counts()

# Paso 5: Ordenar las estaciones en el orden correcto (primavera, verano, otoño, invierno)
season_order = ['Spring', 'Summer', 'Autumn', 'Winter']
encounters_per_season = encounters_per_season.reindex(season_order)

# Paso 6: Visualizar los datos en un gráfico de barras
plt.figure(figsize=(8, 5))
encounters_per_season.plot(kind='bar', color=['#90caf9', '#ffb74d', '#81c784', '#f06292'])
plt.title('Frecuencia de encuentros OVNI por estación', fontsize=14)
plt.xlabel('Estación', fontsize=12)
plt.ylabel('Cantidad de encuentros', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```


Conclusión: El patrón estacional muestra que los avistamientos son más frecuentes en los meses cálidos (verano y otoño) y menos comunes durante los meses más fríos (invierno). Esto podría deberse a factores como la mayor actividad al aire libre en los meses cálidos o condiciones climáticas más favorables para la observación del cielo.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegurarse de que la columna 'Hour' esté en el formato correcto
df_ufo['Hour'] = pd.to_numeric(df_ufo['Hour'], errors='coerce')

# Agrupar por hora y contar la cantidad de avistamientos
hourly_sightings = df_ufo['Hour'].value_counts().sort_index()

# Crear un gráfico de barras para visualizar los resultados
plt.figure(figsize=(10, 6))
plt.bar(hourly_sightings.index, hourly_sightings.values, color='skyblue')
plt.xticks(range(0, 24), labels=range(0, 24))
plt.title("Frecuencia de avistamientos OVNI por hora del día")
plt.xlabel("Hora del día")
plt.ylabel("Cantidad de avistamientos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Tabla de las horas con mayor frecuencia de avistamientos
top_hours = hourly_sightings.sort_values(ascending=False).head(5)
print("Horas con más avistamientos:\n")
print(top_hours)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# 1. Filtrar y preparar los datos
df_ufo_clean = df_ufo.dropna(subset=["Country", "length_of_encounter_seconds"])  # Eliminar filas con datos faltantes
df_ufo_clean["length_of_encounter_seconds"] = pd.to_numeric(df_ufo_clean["length_of_encounter_seconds"], errors="coerce")

# 2. Calcular la duración promedio por país
durations_by_country = df_ufo_clean.groupby("Country")["length_of_encounter_seconds"].mean().sort_values(ascending=False)

# 3. Seleccionar los países con más reportes (opcional, para evitar sesgo por pocos reportes)
top_countries = df_ufo_clean["Country"].value_counts().head(10).index
durations_top_countries = durations_by_country.loc[top_countries]

# 4. Visualizar los resultados
plt.figure(figsize=(10, 6))
sns.barplot(x=durations_top_countries.index, y=durations_top_countries.values, palette="viridis")
plt.title("Duración promedio de encuentros OVNI por país", fontsize=14)
plt.xlabel("País", fontsize=12)
plt.ylabel("Duración promedio (segundos)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Análisis estadístico (opcional, si se desea profundizar)
from scipy.stats import f_oneway

# Datos por país para ANOVA (considera los países con más reportes)
anova_data = [df_ufo_clean[df_ufo_clean["Country"] == country]["length_of_encounter_seconds"]
              for country in top_countries]
anova_result = f_oneway(*anova_data)
print(f"Resultado ANOVA: F={anova_result.statistic}, p={anova_result.pvalue}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo CSV (ajusta la ruta a tu archivo)
df_ufo = pd.read_csv('ufo-sightings.csv')

# Asegurarse de que la columna 'Year' esté en formato numérico
df_ufo['Year'] = pd.to_numeric(df_ufo['Year'], errors='coerce')

# Contar los avistamientos por año
avistamientos_por_ano = df_ufo['Year'].value_counts().sort_index()

# Calcular el promedio móvil para suavizar la serie y ver mejor los picos
avistamientos_movil = avistamientos_por_ano.rolling(window=3).mean()

# Identificar los picos significativos: donde el número de avistamientos supera el promedio móvil
umbral_pico = avistamientos_movil.mean() + 2 * avistamientos_movil.std()  # Umbral definido como 2 desviaciones estándar por encima del promedio

# Años con picos significativos
picos_significativos = avistamientos_movil[avistamientos_movil > umbral_pico]

# Visualizar la distribución de los avistamientos a lo largo del tiempo con el promedio móvil
plt.figure(figsize=(10, 6))
plt.plot(avistamientos_por_ano.index, avistamientos_por_ano, label='Avistamientos por Año', color='blue', marker='o', linestyle='-', linewidth=2)
plt.plot(avistamientos_movil.index, avistamientos_movil, label='Promedio Móvil (3 años)', color='red', linestyle='--', linewidth=2)
plt.scatter(picos_significativos.index, picos_significativos, color='green', label='Picos Significativos', zorder=5)
plt.title('Avistamientos de OVNIs por Año con Picos Significativos')
plt.xlabel('Año')
plt.ylabel('Número de Avistamientos')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir los años con picos significativos
print("Años con picos significativos en los avistamientos de OVNIs:")
print(picos_significativos)
```

Estos picos podrían estar relacionados con el interés social y eventos globales que impulsaron el reporte de OVNIs.

# Evaluación del modelo

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Convertir variables categóricas a numéricas
prueba = pd.get_dummies(prueba, columns=['Season', 'Country_Code'], drop_first=True)

# Seleccionar las columnas de interés
X = prueba[['Year', 'Month', 'Hour', 'length_of_encounter_seconds'] +
           [col for col in prueba.columns if 'Season_' in col or 'Country_Code_' in col]]  # Predictoras
y_latitude = prueba['latitude']  # Objetivo (latitud)
y_longitude = prueba['longitude']  # Objetivo (longitud)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_lat_train, y_lat_test = train_test_split(X, y_latitude, test_size=0.2, random_state=42)
X_train, X_test, y_lon_train, y_lon_test = train_test_split(X, y_longitude, test_size=0.2, random_state=42)

# Crear y entrenar los modelos de regresión
model_latitude = LinearRegression()
model_longitude = LinearRegression()

model_latitude.fit(X_train, y_lat_train)
model_longitude.fit(X_train, y_lon_train)

# Realizar predicciones
y_lat_pred = model_latitude.predict(X_test)
y_lon_pred = model_longitude.predict(X_test)

# Evaluar los modelos
mse_lat = mean_squared_error(y_lat_test, y_lat_pred)
r2_lat = r2_score(y_lat_test, y_lat_pred)

mse_lon = mean_squared_error(y_lon_test, y_lon_pred)
r2_lon = r2_score(y_lon_test, y_lon_pred)

print("Evaluación para latitud:")
print("MSE:", mse_lat)
print("R²:", r2_lat)

print("\nEvaluación para longitud:")
print("MSE:", mse_lon)
print("R²:", r2_lon)
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, classification_report

# Variables para predicción
X = prueba[['Year', 'Month', 'Hour', 'length_of_encounter_seconds'] +
           [col for col in prueba.columns if 'Season_' in col or 'Country_Code_' in col]]  # Predictoras

# Objetivos
y_latitude = prueba['latitude']  # Latitud
y_longitude = prueba['longitude']  # Longitud
y_shape = prueba['UFO_shape']  # Tipo de OVNI

# Convertir UFO_shape a valores numéricos (clasificación)
y_shape_encoded = pd.factorize(y_shape)[0]

# Dividir los datos para regresión y clasificación
X_train, X_test, y_lat_train, y_lat_test = train_test_split(X, y_latitude, test_size=0.2, random_state=42)
X_train, X_test, y_lon_train, y_lon_test = train_test_split(X, y_longitude, test_size=0.2, random_state=42)
X_train_shape, X_test_shape, y_shape_train, y_shape_test = train_test_split(X, y_shape_encoded, test_size=0.2, random_state=42)

# Modelos de Random Forest
model_rf_lat = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf_lon = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf_shape = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenar modelos
model_rf_lat.fit(X_train, y_lat_train)
model_rf_lon.fit(X_train, y_lon_train)
model_rf_shape.fit(X_train_shape, y_shape_train)

# Predicciones
y_lat_pred = model_rf_lat.predict(X_test)
y_lon_pred = model_rf_lon.predict(X_test)
y_shape_pred = model_rf_shape.predict(X_test_shape)

# Evaluar modelos de regresión
mse_lat = mean_squared_error(y_lat_test, y_lat_pred)
r2_lat = r2_score(y_lat_test, y_lat_pred)

mse_lon = mean_squared_error(y_lon_test, y_lon_pred)
r2_lon = r2_score(y_lon_test, y_lon_pred)

# Evaluar modelo de clasificación
print("\nEvaluación para predicción de forma del OVNI (clasificación):")
print(classification_report(y_shape_test, y_shape_pred))

# Mostrar evaluación de regresión
print("\nEvaluación para latitud con Random Forest:")
print("MSE:", mse_lat)
print("R²:", r2_lat)

print("\nEvaluación para longitud con Random Forest:")
print("MSE:", mse_lon)
print("R²:", r2_lon)

# Predicción para nuevos datos
nuevo_dato = [[2024, 3, 22, 3600, 1, 0, 0, 1, 0, 0]]  # Año, Mes, Hora, duración y variables categóricas
lat_pred_new = model_rf_lat.predict(nuevo_dato)
lon_pred_new = model_rf_lon.predict(nuevo_dato)
shape_pred_new = model_rf_shape.predict(nuevo_dato)

# Decodificar la forma del OVNI predicha
shape_pred_label = pd.factorize(y_shape)[1][shape_pred_new[0]]

print("\nPredicción para nuevos datos:")
print(f"Latitud: {lat_pred_new[0]}")
print(f"Longitud: {lon_pred_new[0]}")
print(f"Forma del OVNI: {shape_pred_label}")
```
